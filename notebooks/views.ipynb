```python
from picow_ws2812_core.collections import Text, Group
from picow_ws2812_core import StaticSequence, StaticView
from picow_ws2812_devtools.visualizer import LedWallVisualizer

from picow_ws2812_core.objects import Char, Box
```

```python
ledwall = LedWallVisualizer(46, 46)
```

```python
city = "Milano"
min = -5
max = 10
weather = "Sunny"
```

```python
group1 = Group([
    Text(city, (0, 0, 255), x0=2, y0=2), 
    Text(f"{min} - {max}°C", (255, 255, 255), x0=2, y0=10),
    Text(weather, (255, 255, 255), x0=2, y0=18),
    Box(0, 0, 45, 45, (16, 255, 255))
    ])
```

```python
view1 = StaticView(ncols=46, nrows=46, objects=[group1])
```

```python
ledwall.render_view(view1)
```
